```python
from random import randint
import numpy as np
import itertools
LINES = 10000
TRAINING_LINES = 9600
MAX_LEN = 10
BATCH_SIZE = 16
```

```python
import nltk, string

translator = str.maketrans("","", string.punctuation)
stopwords = set(nltk.corpus.stopwords.words('english'))

def my_tokenizer(s):
#     return [ x for x in nltk.tokenize.word_tokenize(s.lower().translate(translator)) if not x in stopwords ]
    return [ x for x in nltk.tokenize.word_tokenize(s.lower().translate(translator))]
```

```python
with open("data/europarl-v7.es-en.en", encoding="utf8") as f:
    en_lines = [ my_tokenizer(x.strip()) for x in itertools.islice(f, LINES) ]
```

```python
with open("data/europarl-v7.es-en.es", encoding="utf8") as f:
    spa_lines = [ my_tokenizer(x.strip()) for x in itertools.islice(f, LINES) ]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def visualize(lines):
    text = " ".join(" ".join(tokens) for tokens in lines)
    cv = CountVectorizer(min_df=0, stop_words=stopwords, max_features=200)
    counts = cv.fit_transform([text]).toarray().ravel()                                                  
    words = np.array(cv.get_feature_names())
    wordcloud = WordCloud(max_font_size=40).generate(" ".join(words))
    plt.figure(figsize=(20, 8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
visualize(en_lines)
```

```python
visualize(spa_lines)
```

```python
spa_words = set([x for line in spa_lines for x in line] + ["\t"])
```

```python
en_words = set([x for line in en_lines for x in line] + ["\t"])
```

```python
spa_dict = {l: i for i, l in enumerate(spa_words)}
inv_spa_dict = {v: k for k, v in spa_dict.items()}

en_dict = {l: i for i, l in enumerate(en_words)}
inv_en_dict = {v: k for k, v in en_dict.items()}
```

```python
spa_vocab = len(spa_dict)
en_vocab = len(en_dict)
en_window = max([len(line)] for line in en_lines)[0] + 1
spa_window = max([len(line)] for line in spa_lines)[0] + 1
```

```python
def generate_sequence(tokens, dic):
    return [dic[x] for x in tokens]
```

```python
def one_hot_encode(sequence, n_unique):
    encoding = []
    for value in sequence:
        vector = [0 for _ in range(n_unique)]
        vector[value] = 1
        encoding.append(vector)
    return encoding
```

```python
# decode a one hot encoded string
def one_hot_decode(encoded_seq):
    return [np.argmax(vector) for vector in encoded_seq]
```

```python
def generate_case(spa_tokens, en_tokens):
    
    sequence_in = generate_sequence(spa_tokens, spa_dict)
    sequence_in = sequence_in + [0 for _ in range(spa_window - len(sequence_in))]
    sequence_out = generate_sequence(en_tokens, en_dict)
    sequence_out = sequence_out + [0 for _ in range(en_window - len(sequence_out))]
    # one hot encode
    X = one_hot_encode(sequence_in, spa_vocab)
    y = one_hot_encode(sequence_out, en_vocab)
    
    return X , y
```

```python
def get_input_batch(start, size):
    
    encoder_input = np.zeros((size, spa_window, spa_vocab))
    decoder_input = np.zeros((size, en_window + 1, en_vocab))
    
    for i in range(size):
        seq_spa = generate_sequence(spa_lines[start + i], spa_dict)
        seq_en = generate_sequence(en_lines[start + i], en_dict)
        
        for j, idx in enumerate(seq_spa):
            encoder_input[i][j][idx] = 1.

        decoder_input[i][0][en_dict["\t"]] = 1.

        for j, idx in enumerate(seq_en):
            decoder_input[i][j + 1][idx] = 1.
            
    return [encoder_input, decoder_input]
```

```python
def get_output_batch(start, size):
    decoder_output = np.zeros((size, en_window + 1, en_vocab))
    
    for i in range(size):
        seq_en = generate_sequence(en_lines[start + i], en_dict)
        
        for j, idx in enumerate(seq_en):
            decoder_output[i][j][idx] = 1.            
    return decoder_output
```

```python
def generate_data(start, end):
    batch_size = BATCH_SIZE
    while True:
        batch_size = min(BATCH_SIZE , end - start - 1)
        if batch_size == 0:
            break
        yield (get_input_batch(start, batch_size), get_output_batch(start, batch_size))
        start += batch_size
        if start >= end:
            break
```

```python
from keras.models import Sequential, Input, Model
from keras.layers.recurrent import LSTM
from keras.layers import RepeatVector, TimeDistributed, Dense, Bidirectional
from attention_decoder import AttentionDecoder
```

```python
def baseline_model():
    n_units = 150
    # define training encoder
    encoder_inputs = Input(shape=(spa_window, spa_vocab))
    encoder = LSTM(n_units, return_state=True)   
    encoder_outputs, state_h, state_c = encoder(encoder_inputs)
    encoder_states = [state_h, state_c]
    # define training decoder
    decoder_inputs = Input(shape=(en_window + 1, en_vocab))
    decoder_lstm = LSTM(n_units, return_sequences=True, return_state=True)
    decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
    decoder_dense = Dense(en_vocab, activation='softmax')
    decoder_outputs = decoder_dense(decoder_outputs)
    
    model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
    # define inference encoder
    encoder_model = Model(encoder_inputs, encoder_states)
    # define inference decoder
    decoder_state_input_h = Input(shape=(n_units,))
    decoder_state_input_c = Input(shape=(n_units,))
    decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
    decoder_outputs, state_h, state_c = decoder_lstm(decoder_inputs, initial_state=decoder_states_inputs)
    decoder_states = [state_h, state_c]
    decoder_outputs = decoder_dense(decoder_outputs)
    decoder_model = Model([decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states)

    return model, encoder_model, decoder_model
```

```python
print('Encoder-Decoder Model')

model, encoder_model, decoder_model = baseline_model()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit_generator(
    generate_data(0, TRAINING_LINES), 
    steps_per_epoch= int(TRAINING_LINES/BATCH_SIZE),
    epochs = 2,
    verbose = 1, 
    validation_data=generate_data(TRAINING_LINES, LINES),
    validation_steps=int((LINES-TRAINING_LINES)/BATCH_SIZE),
    shuffle=True)
```

```python
def convert_to_input_data(s):
    tokens = my_tokenizer(s.strip())
    inp_data = np.zeros((1, spa_window, spa_vocab))
    for j, c in enumerate(tokens):
        inp_data[0, j, spa_dict[c]] = 1.
    return inp_data
```

```python
def decode_sequence(input_seq):
    input_seq = convert_to_input_data(input_seq)
    
    # Encode the input as state vectors.
    states_value = encoder_model.predict(input_seq)

    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1, en_window + 1, en_vocab))
    # Populate the first character of target sequence with the start character.
    target_seq[0, 0, en_dict['\t']] = 1.

    # Sampling loop for a batch of sequences
    # (to simplify, here we assume a batch of size 1).
    stop_condition = False
    decoded_sentence = ''
    pos = 1
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value)

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])       
        sampled_word = inv_en_dict[sampled_token_index]
        decoded_sentence += sampled_word + " "

        # Exit condition: either hit max length
        # or find stop character.
        if (sampled_word == '\n' or
           len(decoded_sentence) > en_window):
            stop_condition = True

        # Update the target sequence (of length 1).      
        target_seq[0, pos, sampled_token_index] = 1.

        # Update states
        states_value = [h, c]
        pos += 1

    return decoded_sentence
    
```
